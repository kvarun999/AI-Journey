```python
import json
import pandas as pd
from pathlib import Path

current_folder = Path(__file__).parent if "__file__" in locals() else Path.cwd()
data_path = current_folder.parent / "mock_data" / "clinical_trials.json"

#loading mock data 
with open(data_path) as f:
    data = json.load(f)

#converting data into pd
df = pd.DataFrame(data["trials"])
print(df)
```

```python
active_trials = df[df["status"] == "Recruiting"]
print(active_trials)
```
