```python
import json
import pandas as pd

with open("../mock_data/market_data.json") as f:
    data = json.load(f)

df_market = pd.DataFrame(data["therapy_areas"])
df_market
```

```python
high_growth = df_market[df_market["cagr_percent"] > 8]
high_growth
```

```python
attractive = df_market[(df_market["cagr_percent"] > 7) & (df_market["competition_level"] == "Low")]
attractive
```

```python
competition_score = {"Low": 3, "Medium": 2, "High": 1}

df_market["competition_score"] = df_market["competition_level"].map(competition_score)
df_market["opportunity_score"] = (
    df_market["competition_score"] * df_market["cagr_percent"] * df_market["market_size_usd_bn"]
)
df_market.sort_values("opportunity_score", ascending=False)
```

```python
top = df_market.sort_values("opportunity_score", ascending=False).iloc[0]

summary = f"""
Market Opportunity Summary

Top Repurposing Candidate:
Therapy Area: {top['therapy']}
Market Size: ${top['market_size_usd_bn']}B
Growth Rate: {top['cagr_percent']}% CAGR
Competition: {top['competition_level']}

Rationale:
High growth combined with manageable competition
indicates strong commercial potential.
"""

print(summary)
```
