```python
import json
import pandas as pd
from datetime import datetime

current_year = datetime.now().year

with open("../mock_data/patents.json") as f:
    data = json.load(f)

df_patents = pd.DataFrame(data["patents"])
print(df_patents)
```

```python
active = df_patents[df_patents["status"] == "Active"]
print(active)

expired = df_patents[df_patents["status"] == "Expired"]
print(expired)
```

```python
df_patents["fto_risk"] = df_patents["expiry_year"] > current_year
df_patents
```

```python
opportunity_markets = df_patents[df_patents["fto_risk"] == False]["country"]
opportunity_markets.tolist()
```

```python
summary = f"""
Patent Landscape - Metformin

Active patents found in: {', '.join(active['country'])}
Patent-free regions: {', '.join(opportunity_markets)}

Conclusion:
Metformin shows freedom-to-operate in select markets, enabling
repurposing opportunities without IP infringement.
"""

print(summary)
```
